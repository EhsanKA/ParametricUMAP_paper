### Load dataset

```python
from sklearn.datasets import make_moons
```

```python
X_train, Y_train = make_moons(1000, random_state=0, noise=0.1)
X_train_flat = X_train
```

```python
X_test, Y_test = make_moons(1000, random_state=1, noise=0.1)
X_test_flat = X_test
```

### Train model

```python
from umap import UMAP
```

```python
embedder = UMAP(verbose=True)
z_umap = embedder.fit_transform(X_train_flat)
```

```python
fig, ax = plt.subplots( figsize=(8, 8))
sc = ax.scatter(
    z_umap[:, 0],
    z_umap[:, 1],
    c=Y_train.astype(int)[:len(z_umap)],
    cmap="tab10",
    s=0.1,
    alpha=0.5,
    rasterized=True,
)
ax.axis('equal')
ax.set_title("UMAP in Tensorflow embedding", fontsize=20)
plt.colorbar(sc, ax=ax);
```

### Save output

```python
import os
import pickle
```

```python
from tfumap.paths import ensure_dir, MODEL_DIR
```

```python
output_dir = MODEL_DIR/'projections'/ 'moons' / 'umap-learn'
ensure_dir(output_dir)
```

```python
with open(os.path.join(output_dir, "model.pkl"), "wb") as output:
    pickle.dump(embedder, output, pickle.HIGHEST_PROTOCOL)
```
