### load packages

```python
from tfumap.umap import tfUMAP
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tqdm.autonotebook import tqdm
import umap
import pandas as pd
```

### Load dataset

```python
import requests
import json
url = "https://raw.githubusercontent.com/duhaime/umap-zoo/03819ed0954b524919671a72f61a56032099ba11/data/json/bison.json"
animal = np.array(json.loads(requests.get(url).text)['3d'])
np.shape(animal)
```

```python
fig, ax = plt.subplots()
ax.scatter(animal[:,2], animal[:,1], s = 1, c = animal[:,0], alpha = 0.1)
ax.axis('equal')
```

```python
X_train = animal
Y_train = animal[:, 2]
X_train_flat = X_train
```

### Create model and train

```python
embedder = tfUMAP(
    n_neighbors=150,
    direct_embedding=True,
    verbose=True,
    #training_epochs=100,
)
```

```python
z = embedder.fit_transform(X_train_flat)
```

### Plot model output

```python
fig, ax = plt.subplots( figsize=(8, 8))
sc = ax.scatter(
    z[:, 0],
    z[:, 1],
    c=Y_train.astype(int)[:len(z)],
    cmap="tab10",
    s=0.1,
    alpha=0.5,
    rasterized=True,
)
ax.axis('equal')
ax.set_title("UMAP in Tensorflow embedding", fontsize=20)
plt.colorbar(sc, ax=ax);
```

### View loss

```python
from tfumap.umap import retrieve_tensors
import seaborn as sns
```

```python
loss_df = retrieve_tensors(embedder.tensorboard_logdir)
loss_df[:3]
```

```python
ax = sns.lineplot(x="step", y="val", hue="group", data=loss_df[loss_df.variable=='umap_loss'])
ax.set_xscale('log')
```
