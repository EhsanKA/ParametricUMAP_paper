```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tqdm.autonotebook import tqdm
import umap
import pandas as pd
```

### Load dataset

```python
dataset = 'bison'
```

```python
import requests
import json
url = "https://raw.githubusercontent.com/duhaime/umap-zoo/03819ed0954b524919671a72f61a56032099ba11/data/json/bison.json"
animal = np.array(json.loads(requests.get(url).text)['3d'])
np.shape(animal)
```

```python
fig, ax = plt.subplots()
ax.scatter(animal[:,2], animal[:,1], s = 1, c = animal[:,0], alpha = 0.1)
ax.axis('equal')
```

```python
X_train = animal
Y_train = animal[:, 2]
X_train_flat = X_train
```

### Create model and train

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
z = pca.fit_transform(X_train_flat)
```

### plot output

```python
fig, ax = plt.subplots( figsize=(8, 8))
sc = ax.scatter(
    z[:, 0],
    z[:, 1],
    c=Y_train,
    #cmap="tab10",
    s=0.1,
    alpha=0.5,
    rasterized=True,
)
ax.axis('equal')
ax.set_title("PCA embedding", fontsize=20)
plt.colorbar(sc, ax=ax);
```

### Save model

```python
import os
import pickle
```

```python
from tfumap.paths import ensure_dir, MODEL_DIR
```

```python
output_dir = MODEL_DIR/'projections'/ dataset / 'PCA'
ensure_dir(output_dir)
```

```python
with open(os.path.join(output_dir, "model.pkl"), "wb") as output:
    pickle.dump(pca, output, pickle.HIGHEST_PROTOCOL)
```

```python
np.save(output_dir / 'z.npy', z)
```

## tsne

```python
from openTSNE import TSNE
```

```python
tsne = TSNE(
    n_components = 2,
    perplexity=2000,
)
```

```python
embedding_train = tsne.fit(X_train_flat)
```

```python
z = np.array(embedding_train)
```

```python
fig, ax = plt.subplots( figsize=(8, 8))
sc = ax.scatter(
    z[:, 0],
    z[:, 1],
    c=Y_train,
    #cmap="tab10",
    s=0.1,
    alpha=0.5,
    rasterized=True,
)
ax.axis('equal')
ax.set_title("PCA embedding", fontsize=20)
plt.colorbar(sc, ax=ax);
```
